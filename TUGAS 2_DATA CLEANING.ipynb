## Missing Value

```python
np.sum(data.isnull())
```

```python
data.isnull().sum().sum()
```

```python
data['occupation'] = data.fillna(data['occupation'].value_counts().index[0])
data['native-country'] = data.fillna(data['native-country'].value_counts().index[0])
data.isnull().any()
```

## Encode Labels

```python
from sklearn.preprocessing import LabelEncoder
```

```python
data['sex'] = LabelEncoder().fit_transform(data['sex'])
data['race'] = LabelEncoder().fit_transform(data['race'])
data.head()
```

## Standardize

```python
from sklearn.preprocessing import StandardScaler
```

```python
data['age'] = StandardScaler().fit(data[['age']]).transform(data[['age']])
data['hours per week'] = StandardScaler().fit(data[['hours per week']]).transform(data[['hours per week']])
data.head()
```

## Normalize

```python
from sklearn.preprocessing import Normalizer
```

```python
data['age'] = Normalizer().fit(data[['age']]).transform(data[['age']])
data['hours per week'] = Normalizer().fit(data[['hours per week']]).transform(data[['hours per week']])
data.head()
```
